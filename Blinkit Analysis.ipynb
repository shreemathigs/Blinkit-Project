BUSINESS REQUIREMENTS

KPI Requirements

```python
#Total Sales
total_sales=df['Sales'].sum()

#Average Sales
avg_sales=df['Sales'].mean()

#No of items sold
no_of_items_sold=df['Sales'].count()

#Average Customer Rating
avg_customer_rating=df['Rating'].mean()

#Display
print(f"Total Sales: ${total_sales:,.0f}")
print(f"Average Sales: ${avg_sales:,.0f}")
print(f"No of items sold: {no_of_items_sold:,.0f}")
print(f"Average Customer Rating: {avg_customer_rating:,.0f}")
```

CHART REQUIREMENTS

Total Sales by Fat Content

```python
sales_by_fat= df.groupby('Item Fat Content')['Sales'].sum()
plt.pie(sales_by_fat,labels=sales_by_fat.index,autopct='%.1f%%',startangle=90)
plt.title('Sales by fat content')
plt.show()
```

Total Sales by Itemtype

```python
sales_by_itemtype = df.groupby('Item Type')['Sales'].sum().sort_values(ascending=False)
print (sales_by_itemtype)

bars = plt.bar (sales_by_itemtype.index, sales_by_itemtype.values)
plt.title('Total Sales by Itemtype')
plt.xlabel('Item Type')
plt.ylabel('Total Sales')
plt.xticks(rotation=90)
plt.show()
```

Fat Content by Outlet for Total Sales:


```python
fat_content_by_outlet = df.groupby(['Outlet Location Type','Item Fat Content'])['Sales'].sum().unstack()
print(fat_content_by_outlet)


fat_content_by_outlet.plot(kind='bar', stacked=True, figsize=(8, 6))
plt.title('Fat Content by Outlet for Total Sales')
plt.xlabel('Outlet Location Type')
plt.ylabel('Total Sales')
plt.show()

```

Total Sales by Outlet Establishment


```python
sales_by_outlet = df.groupby('Outlet Establishment Year')['Sales'].sum()
print(sales_by_outlet)

plt.figure(figsize =(10,6))
plt.plot(sales_by_outlet.index, sales_by_outlet.values, marker='o', linestyle='-')
plt.title('Total Sales by Outlet Establishment')
plt.xlabel('Outlet Establishment Year')
plt.ylabel('Total Sales')
plt.show()

```

Sales By Outlet Size

```python
outlet_size = df.groupby('Outlet Size')['Sales'].sum()
print(outlet_size)

plt.pie(outlet_size,labels=outlet_size.index, autopct='%.2f%%', startangle=90)
plt.xlabel('Outlet Size')
plt.ylabel('Total Sales')
plt.title('Sales by Outlet Size')
plt.show()
```

Sales by Outlet Location

```python
outlet_location = df.groupby('Outlet Location Type')['Sales'].sum().reset_index()
outlet_location = outlet_location.sort_values('Sales', ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x='Sales', y='Outlet Location Type', data=outlet_location, palette='Blues_d')

plt.xlabel('Total Sales')
plt.ylabel('Outlet Location Type')
plt.title('Total Sales by Outlet Location Type')
plt.tight_layout()
plt.show()
```
